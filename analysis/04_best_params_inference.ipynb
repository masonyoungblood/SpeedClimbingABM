This notebook runs training on the CPU. I tried setting up GPU acceleration on my M1 Mac for training and it was at least two orders of magnitude slower.

```python
#import necessary libraries
import tensorflow as tf
import bayesflow as bf
import numpy as np
import json
```

```python
#load in training data and format
with open("data_and_output/03_best_params/train_data.json") as f:
    l = json.load(f)
train_data = {}
train_data["prior_non_batchable_context"] = None
train_data["prior_batchable_context"] = None
train_data["prior_draws"] = np.array(l[0], dtype = float)
train_data["sim_non_batchable_context"] = None
train_data["sim_batchable_context"] = None
train_data["sim_data"] = np.array(l[1], dtype = float)

#load in testing data and format
with open("data_and_output/03_best_params/test_data.json") as f:
    l = json.load(f)
test_data = {}
test_data["prior_non_batchable_context"] = None
test_data["prior_batchable_context"] = None
test_data["prior_draws"] = np.array(l[0], dtype = float)
test_data["sim_non_batchable_context"] = None
test_data["sim_batchable_context"] = None
test_data["sim_data"] = np.array(l[1], dtype = float)

#load in observed data and format
with open("data_and_output/03_best_params/obs_data.json") as f:
    l = json.load(f)
obs_data = {}
obs_data["prior_non_batchable_context"] = None
obs_data["prior_batchable_context"] = None
obs_data["prior_draws"] = None
obs_data["sim_non_batchable_context"] = None
obs_data["sim_batchable_context"] = None
obs_data["sim_data"] = np.array(l, dtype = float)
```

```python
#set number of parameters, epochs, batch size, and learning rate
num_params = 13
epochs = 200
batch_size = 32
learning_rate = 0.0005
```

```python
#initialize summary network, inference network, and amortized posterior
#default settings live here: https://github.com/stefanradev93/BayesFlow/blob/master/bayesflow/default_settings.py
summary_net = bf.networks.TimeSeriesTransformer(input_dim = 380, num_dense_fc = 4, summary_dim = 256, dense_settings = {"units": 512, "activation": "relu"}, template_type = "gru")
inference_net = bf.networks.InvertibleNetwork(num_params = num_params, num_coupling_layers = 10, coupling_design = "affine", coupling_settings = {"dropout_prob": 0.05, "num_dense": 4, "dense_args": {"units": 128, "activation": "relu"}})
amortized_posterior = bf.amortizers.AmortizedPosterior(inference_net, summary_net)
```

```python
#create trainer
trainer = bf.trainers.Trainer(amortizer = amortized_posterior, default_lr = learning_rate, memory = True, checkpoint_path = "data_and_output/04_best_params_inference/checkpoints")
```

```python
#setup custom optimizer that runs faster on apple silicon
#adapted from _setup_optimizer and train_offline in trainers.py, and defaults in OPTIMIZER_DEFAULTS in default_settings.py
schedule = tf.keras.optimizers.schedules.CosineDecay(learning_rate, epochs*bf.trainers.SimulationDataset(train_data, batch_size).num_batches, name = "lr_decay")
optimizer = tf.keras.optimizers.legacy.Adam(schedule, global_clipnorm = 1)
```

```python
#run training and save
history = trainer.train_offline(simulations_dict = train_data, epochs = epochs, batch_size = batch_size, optimizer = optimizer)
trainer._save_trainer("data_and_output/04_best_params_inference/model")
```

```python
#load trained network
trainer.load_pretrained_network()
```

```python
#sample 100 posterior draws per simulation run in the test set
test_set = trainer.configurator(test_data)
test_posteriors = amortized_posterior.sample(test_set, n_samples = 100)
```

```python
#plot and save the model loss
loss_plot = bf.diagnostics.plot_losses(history, train_color = "black", lw_train = 1)
loss_plot.savefig("data_and_output/04_best_params_inference/loss_plot.pdf")
```

```python
#plot recovery of parameter values
recovery_plot = bf.diagnostics.plot_recovery(test_posteriors, test_set["parameters"],
                                             param_names = ["Innovation Probability", "Copying Probability", "Number of Top Climbers",
                                                            "Minimum Improvement (M)", "Minimum Improvement (W)",
                                                            "Rate of Improvement (M)", "Rate of Improvement (W)",
                                                            "Innovation x Year", "Innovation x Climbing Time", "Innovation x Population Size",
                                                            "Copying x Year", "Copying x Climbing Time", "Copying x Population Size"], color = "black", add_corr = False)
recovery_plot.savefig("data_and_output/04_best_params_inference/recovery_plot.pdf")
```

```python
#results of simulation-based calibration, with explanation here: https://arxiv.org/pdf/1804.06788.pdf
sbc_plot = trainer.diagnose_sbc_histograms(plot_args = {"param_names": ["Innovation Probability", "Copying Probability", "Number of Top Climbers", 
                                                                        "Minimum Improvement (M)", "Minimum Improvement (W)",
                                                                        "Rate of Improvement (M)", "Rate of Improvement (W)",
                                                                        "Innovation x Year", "Innovation x Climbing Time", "Innovation x Population Size",
                                                                        "Copying x Year", "Copying x Climbing Time", "Copying x Population Size"],
                                                                        "num_bins": 10, "hist_color": "#6e6e6e"})
sbc_plot.savefig("data_and_output/04_best_params_inference/sbc_plot.pdf")
```

```python
#plot the latent space in the model
latent_plot = trainer.diagnose_latent2d(plot_args = {"height": 1.2, "color": "#6e6e6e"})
latent_plot.savefig("data_and_output/04_best_params_inference/latent_plot.pdf")
```

```python
#get and save 10000 posterior predictions for each parameter based on the real data
obs_set = trainer.configurator(obs_data)
posterior_predictions = amortized_posterior.sample(obs_set, n_samples = 10000)

with open("data_and_output/04_best_params_inference/posterior_predictions.json", "w") as f:
    json.dump(posterior_predictions.tolist(), f)
```
